# Converting to TF-IDF

```python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB #multinomial used for bias and discrete
from sklearn.feature_extraction.text import TfidfVectorizer
X_train=np.array(["This movie is absolutely fantastic,i loved it!",
                  "The plot was terrible and the acting was wooden",
                  "A stunning masterpiece with brilliant performance.",
                  "I hated every minute of this boring film",
                  "the director did an amazing job,highly recommended",
                  "The script was weak and the characters were poorly developed",
                  "What a wonderful and heartwarming story",
                  "poor cinematogrophy and predictable ending",
                  "i was on the edge of my seat the entire time",
                  "A complete waste of time,nothing original"])
y_train=np.array(["positive","negative","positive","negative","positive","negative",
"positive","negative","positive","negative"])
#passing to model
clf=Pipeline([("tfidf",TfidfVectorizer()),("naive_bayes",MultinomialNB())]) #pipeline used for sequential root output passing from one to another
clf.fit(X_train,y_train)
```

```python
new=clf.predict(["this was an amazing movie"]) #prediction
new
```

```python
new=clf.predict(["very poor movie bored"])
new
```
