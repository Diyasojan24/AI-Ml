```python
from transformers import pipeline
text_generator=pipeline("text-generation",model="distilgpt2")
generated_text=text_generator(
    "today is a rainy day in london",
    truncation=True,
    num_return_sequences=1,
    max_new_tokens=10
)
print("generated_text:\n",generated_text)
```

Task 3: question answering

```python
from transformers import pipeline
qa_model=pipeline("question-answering")
question="what is my job?"
context="I am  developing AI models with python"
print(qa_model(question=question,context=context))
```

```python
from transformers import pipeline

# Load a generative model, for example, 't5-large' or another LLM
# Be mindful of model size and available hardware
qa_pipeline = pipeline("text-generation", model="t5-base") # T5 is good for text generation tasks

question = "What is the capital of France?"

# The model generates an answer based on its pre-trained knowledge
answer = qa_pipeline(f"Question: {question} Answer:")
print(answer[0]['generated_text'])
```

```python
from transformers import pipeline
generator=pipeline("text-generation",model="openai-community/gpt2")
question="who is the president of US?"
prompt=f"Question: {question}\nAnswer:"
result=generator(prompt,max_new_tokens=20,num_return_sequences=1)
print(result[0]['generated_text'])
```
