#LLM Application

```python
def generate_text(prompt):
  result=ollama.generate(model='llama2,prompt=prompt')
  return result['response']
```

```python
def question_answer(question):
  prompt=f'Answer the question concisely : {question}'
  result=generate_text(prompt)
  return result['response']
```

```python
def summarize(paragraph):
  prompt=f'summarize the below paragraph into 2 sentence \n : {paragraph}'
  result=generate_text(prompt)
  return result['response']
```

```python
def lang_transalte(sentence,language):
  prompt=f'convert the below sentence \n {sentence} in to {language} language'
  result=generate_text(prompt)
  return result['response']
```

```python
print("LLM Application NLP tasks")
print("1.generate text")
print("2.quetsion answering")
print("3.summary of paragraph")
print("4.language translation")
print("enter your choice")
choice=input("1/2/3/4")
if choice=='1':
  prompt=input("enter your prompt here")
  result=generate_text(prompt)
elif choice=='2':
  question=input("enter you question here")
  result=generate_text(question)
elif choice=='3':
  paragraph=input("enter your paragraph to summarize")
  result=generate_text(paragraph)
elif choice=='4':
  sentence=input("enter the sentence to translate")
  language=input("enter the language")
  result=lang_transalte(sentence,language)
else:
  result="wrong choice !!!"
print("result:",result)
```
